```python
import numpy as np
import matplotlib.pyplot as plt

seconds = 1
fs=1000
t = np.arange(0, seconds, 1.0/fs)

def gen_signal(freq, A=1, phi=0):
    return A*np.sin(2*np.pi*freq*t + phi)

def plot_signal(signal, title=''):
    plt.figure(figsize=(15, 3))
    limit = max(signal)*1.1
    limit = 1 if limit < 1 else limit
    plt.ylim(top=limit, bottom=-limit)
    plt.xlabel('time')
    plt.ylabel('amplitude')
    plt.title(title)
    plt.plot(t, signal, scaley=False)
```

### Exercise 1

Generate 5 senoidal signals:
- 1Hz
- 5Hz
- 10Hz
- 50Hz
- 100Hz

```python
frequencies = [1, 5, 10, 50, 100]
signals = [gen_signal(f) for f in frequencies]
titles = [f'Senoid - {f}Hz' for f in frequencies]
for signal, title in zip(signals, titles):
    plot_signal(signal, title)
```

### Exercise 2
Plot sum of all signals

```python
plot_signal(sum(signals), 'Sum of all signals')
```
