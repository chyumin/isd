### Show rasterplot and histogram for the data stored in:
rat4-preto-WD-20150316-60min-239trials-92PC-SPK.mat

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io import loadmat
from scipy import signal

fs = 1000
window = 5
fc = 500
filter_order = 10

event_data = loadmat('event.mat', squeeze_me=True)
events = event_data.get('event')

data_file = 'rat4-preto-WD-20150316-60min-239trials-92PC-SPK.mat'
data = loadmat(data_file)

single_neuron_spikes = np.squeeze(data.get('VI32aR'))
```

```python
count = 1
all_event_spikes = []
for event in events:
    event_spikes = [ts for ts in single_neuron_spikes if ts >= event-4 and ts <= event+4]
    event_spikes = [spike-event for spike in event_spikes]
    all_event_spikes.append(event_spikes)
    plt.plot(event_spikes, [count]*len(event_spikes), '|')
    count += 1
```

```python
_, _, _ = plt.hist(np.concatenate(all_event_spikes), bins=400)
```
