```python
import numpy as np
import matplotlib.pyplot as plt

N = 30

A = 1  # Signal amplitude
f = 1  # Fundamental frequency
w = 2*np.pi*f
fs = 1000
at = 5; # Aquisition time
t = np.arange(0, at, 1/fs)  # Time vector


bn = np.zeros(N)
pn = np.zeros((N, len(t)))

for n in range(N):

    if n%2:
        bn[n] = 4*A/(n*np.pi)

    pn[n, :] = bn[n]*np.sin(n*w*t)

plt.plot(sum(pn))
```
